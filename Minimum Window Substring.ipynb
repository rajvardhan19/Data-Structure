# Minimum Window Substring
Given two strings s and t of lengths m and n respectively, return the minimum window 
substring  of s such that every character in t (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.

 
**Example 1:**
```
Input: s = "ADOBECODEBANC", t = "ABC"
Output: "BANC"
Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.
```
**Example 2:**
```
Input: s = "a", t = "a"
Output: "a"
Explanation: The entire string s is the minimum window.
```

**Example 3:**
```
Input: s = "a", t = "aa"
Output: ""
Explanation: Both 'a's from t must be included in the window.
Since the largest window of s only has one 'a', return empty string.
```

--> **Constraints:**
```
m == s.length
n == t.length
1 <= m, n <= 105
s and t consist of uppercase and lowercase English letters.
```


Link --> https://leetcode.com/problems/minimum-window-substring/

````python
class Solution(object):
    def minWindow(self, s, t):
        if not s or not t:
            return ""

        # Count the characters in string t
        target_counts = Counter(t)

        # Initialize variables for the sliding window
        left, right = 0, 0
        min_window = float('inf')
        min_window_start = 0

        # Number of characters left to match in string t
        required_chars = len(target_counts)

        while right < len(s):
            # Expand the window
            if s[right] in target_counts:
                target_counts[s[right]] -= 1
                if target_counts[s[right]] == 0:
                    required_chars -= 1

            # Check if all characters are matched
            while required_chars == 0:
                # Update the minimum window if smaller
                if right - left < min_window:
                    min_window = right - left
                    min_window_start = left

                # Contract the window from the left
                if s[left] in target_counts:
                    target_counts[s[left]] += 1
                    if target_counts[s[left]] > 0:
                        required_chars += 1

                left += 1

            # Expand the window from the right
            right += 1

        # Check if a valid window was found
        if min_window == float('inf'):
            return ""
        else:
            return s[min_window_start:min_window_start + min_window + 1]
        
````
