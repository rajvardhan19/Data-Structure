# Problem Name: Longest Increasing Subsequence

Given an integer array nums, return the number of all the arithmetic subsequences of nums.

A sequence of numbers is called arithmetic if it consists of at least three elements and if the difference between any two consecutive elements is the same.

- For example, ```[1, 3, 5, 7, 9], [7, 7, 7, 7],``` and ```[3, -1, -5, -9]``` are arithmetic sequences.
- For example, ```[1, 1, 2, 5, 7]``` is not an arithmetic sequence.

A subsequence of an array is a sequence that can be formed by removing some elements (possibly none) of the array.

- For example, ```[2,5,10]``` is a subsequence of ```[1,2,1,2,4,1,5,10]```.
The test cases are generated so that the answer fits in 32-bit integer.
 

**Example 1:**
```
Input: nums = [2,4,6,8,10]
Output: 7
Explanation: All arithmetic subsequence slices are:
[2,4,6]
[4,6,8]
[6,8,10]
[2,4,6,8]
[4,6,8,10]
[2,4,6,8,10]
[2,6,10]
```



**Example 2:**
```
Input: nums = [7,7,7,7,7]
Output: 16
Explanation: Any subsequence of this array is arithmetic.
```


 
**Constraints:**
```
1  <= nums.length <= 1000
-231 <= nums[i] <= 231 - 1
```

Link --> https://leetcode.com/problems/arithmetic-slices-ii-subsequence/

````python
class Solution(object):
    def numberOfArithmeticSlices(self, nums):
        total_count = 0
        dp = [{} for _ in nums]

        for i in range(len(nums)):
            for j in range(i):
                diff = nums[i] - nums[j]
                count_j = dp[j].get(diff, 0)
                count_i = dp[i].get(diff, 0)

                dp[i][diff] = count_i + count_j + 1
                total_count += count_j

        return total_count
        
````

1. Initialization:

- total_count keeps track of the total count of arithmetic subsequences.
- dp is a list of dictionaries. Each element dp[i] represents the dictionary for the ith element in the nums array. This dictionary stores the counts of subsequences ending at the current index with specific differences.
2. Loop through the array:

- The outer loop (for i in range(len(nums))) iterates through each element of the nums array.

- The inner loop (for j in range(i)) iterates over the elements before the current ith element.

3. Calculate the difference:

- diff is calculated as the difference between the current element nums[i] and the previous elements nums[j].

4. Count subsequences:

- count_j retrieves the count of subsequences ending at index j with the common difference diff.

- count_i retrieves the count of subsequences ending at index i with the common difference diff.

- dp[i][diff] updates the count of subsequences ending at index i with the common difference diff, considering both the counts from indices i and j. Adding 1 is for considering the new subsequence formed by including the current element.

- total_count accumulates the count of valid subsequences by adding the count of subsequences ending at index j (which form a valid subsequence with the current ith element).

5. Return the total count:

- Finally, the function returns the total_count, which represents the total number of arithmetic subsequences found in the nums array.
