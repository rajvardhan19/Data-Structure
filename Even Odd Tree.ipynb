# Find Bottom Left Tree Value

A binary tree is named Even-Odd if it meets the following conditions:

- The root of the binary tree is at level index 0, its children are at level index 1, their children are at level index 2, etc.
- For every even-indexed level, all nodes at the level have odd integer values in strictly increasing order (from left to right).
- For every odd-indexed level, all nodes at the level have even integer values in strictly decreasing order (from left to right).
Given the root of a binary tree, return true if the binary tree is Even-Odd, otherwise return false.
 
**Example 1:**
```
Input: root = [1,10,4,3,null,7,9,12,8,6,null,null,2]
Output: true
Explanation: The node values on each level are:
Level 0: [1]
Level 1: [10,4]
Level 2: [3,7,9]
Level 3: [12,8,6,2]
Since levels 0 and 2 are all odd and increasing and levels 1 and 3 are all even and decreasing, the tree is Even-Odd.
```
**Example 2:**
```
Input: root = [5,4,2,3,3,7]
Output: false
Explanation: The node values on each level are:
Level 0: [5]
Level 1: [4,2]
Level 2: [3,3,7]
Node values in level 2 must be in strictly increasing order, so the tree is not Even-Odd.
```
**Example 3:**
```
Input: root = [5,9,1,3,5,7]
Output: false
Explanation: Node values in the level 1 should be even integers.
```


--> **Constraints:**
```
The number of nodes in the tree is in the range [1, 105].
1 <= Node.val <= 106
```


Link --> https://leetcode.com/problems/even-odd-tree/

```
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isEvenOddTree(self, root):
        # Initialize a queue to store the nodes at each level, 
        # starting with the root as the first level
        level = [root]
        # Initialize a boolean flag to check if we are on an even or odd level
        evenLevel = True
        # Iterate over the levels
        while level:
            # Initialize the previous value
            # If we are on an even level, it should be less than every value
            # If we are on an odd level, it should be greater than every value
            prev_val = float('-inf') if evenLevel else float('inf')
            # Check each node in the current level
            for _ in range(len(level)):
                # Pop the first node in the level
                node = level.pop(0)
                # If we are on an even level, the values should be increasing and odd
                if evenLevel:
                    if node.val <= prev_val or node.val % 2 == 0:
                        return False
                # If we are on an odd level, the values should be decreasing and even
                else:
                    if node.val >= prev_val or node.val % 2 == 1:
                        return False
                # Update the previous value
                prev_val = node.val
                # Add the children of the current node to the level
                if node.left:
                    level.append(node.left)
                if node.right:
                    level.append(node.right)
            # Flip the boolean flag to switch to the other type of level
            evenLevel = not evenLevel
        # If we reach the end of the tree without returning, the tree meets the conditions
        return True
        
```
